```python
X = df.ix[:, df.columns != "Class"]
Y = df.ix[:, df.columns == "Class"]

# Standarizing the features
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
# Shrink size of data to see if memory issues go away
# Reduce to 50% of original
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_std, Y, test_size=0.5, random_state=42)
```

```python
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_train)
```

```python
plt.figure(figsize=(10,5))
colours = "rgb"
for i in range(pca_components.shape[0]):
    plt.text(pca_components[i, 0], 
             pca_components[i, 1], 
             str(Y_test.Class[i]),
             color=colours[int(Y_test.Class[i])],
             fontdict={'weight': 'bold', 'size': 50}
        )
        

plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```
