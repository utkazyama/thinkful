```python
import pandas as pd
import numpy as np
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Data Exploration & Feature Selection

```python
# I'll use the wine quality dataset from UC Irvaine data repository
df = pd.read_csv("winequality-red.csv", delimiter=';')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

Let's check weather there are missing values.

```python
df.isnull().sum()
```

There is no missing value in this dataset. Next, we will check the distributions of the features.

```python
features = df.loc[:, ~df.columns.isin(["target"])]

plt.figure(figsize=(15,20))

i =1
for column in features.columns:
    plt.subplot(5, 4, i)
    sns.distplot(features[column])
    i+=1
    
plt.tight_layout()
plt.show()
```

Only density and pH are close to normal distributions. Also, there are sharp discontinutities in quality. We don't have continuous data. No wine has a 3.5 rating, for example. 

Now, let's check the correlation between features themselves? Strong correlations between variables can throw off our models.

# quality is not continuous, so it will be categorical


```python
plt.figure(figsize=(7,5))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, vmin=0.0)
plt.show()
```

There are some strong correlationships.

 - citric acid x fixed acidity
 - density x fixed acidity
 - citric acid x density

```python
# Inspect variable correlation for feature selection
corr_matrix = df.corr()
print(corr_matrix)
```

Based on the correlation matrix, I will select only features with >0.20 or -0.20> to the target variable "quality".

```python
data = df.loc[:, ["volatile acidity", "citric acid", "sulphates", "alcohol", "quality"]]

corr_matrix_data = data.corr()
print(corr_matrix_data)
```

```python
plt.figure(figsize=(7,5))
corr_matrix_features = df.loc[:, ["volatile acidity", "citric acid", "sulphates", "alcohol"]].corr()
sns.heatmap(corr_matrix_features, vmin=0.0)
plt.show()
```

Since citric acid and sulphates has a strong correlationship, I will remove citric acid.

## Feature Engineering

```python
x = data.loc[:, ["volatile acidity",  "citric acid", "sulphates", "alcohol"]]
y = data.loc[:, ["quality"]]


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state=43)
```

```python
from sklearn import linear_model

lm = linear_model.LinearRegression()
lm.fit(x_train, y_train)

print('Regression score - training data:', lm.score(x_train, y_train))
print('Regression score - test data:', lm.score(x_test, y_test))
```
